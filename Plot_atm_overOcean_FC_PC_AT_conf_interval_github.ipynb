# Multiply by appropriate constants

```python
def mult_const(Tum,Tvm,vert_adv,Ta2,Tana,Tdel2T,Tdel4T,Tn1,TTo,T2o):
    
    Tum_plot = -Tum * atnorm * 2
    Tvm_plot = -Tvm * atnorm * 2
    
    Had_plot = - (Tum + Tvm) * atnorm * 2
    
    vrt_plot = -vert_adv * atnorm * 2
    
    rah = (1/(rho*Cp)) * (Dmdown*T2o + D0up*TTo) * atnorm * 2
    
    slh = (1/(rho*Cp)) * (lamda*TTo - lamda*T2o) * atnorm * 2
    
    Fmup = -(1/(rho*Cp)) * (Bmup * Tana + Dmup * Ta2) * atnorm * 2
    
    F1down = -(1/(rho*Cp)) * (A11down * Tn1 + B1down * Tana) * atnorm  * 2
    
    Fme_plot = -(phi_m/(rho*Cp)) * (Tana) * atnorm  * 2
    
    ath = Fmup + F1down + Fme_plot
    
    Tdel2T_plot = K2 * Tdel2T * atnorm * 2
    
    Tdel4T_plot = - K4 * Tdel4T * atnorm * 2
    
    Tde = Tdel2T_plot + Tdel4T_plot
    
    sum_terms = np.sum((Tum_plot,Tvm_plot,Fmup,F1down,vrt_plot,slh,rah,Fme_plot,Tdel2T_plot,Tdel4T_plot),axis=0)
    
    return Had_plot,vrt_plot,slh,rah,ath,Tde,sum_terms


```
